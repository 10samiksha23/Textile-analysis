# **line Graph**

```python
import matplotlib.pyplot as plt

pd.pivot_table(
    df,
    index="source",
    values="ps_value_stuivers",
    aggfunc="mean"
).plot(kind="line")

plt.xticks(rotation=45)  # Rotate x-axis tick labels by 45 degrees
plt.tight_layout()       # Adjust layout to prevent label cutoff
plt.show()
```

# **pie chart**

```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Belvliet','Sem Galey','Leusden','Beekestijn']
sizes = [40, 25, 20, 15]  # Values for each label
colors = ['gold', 'lightblue', 'lightcoral', 'green']
explode = (0.1, 0, 0, 0)  # "explode" the first slice

# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%',
        startangle=140, explode=explode, shadow=True)

plt.title('Sales Distribution of Products')
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle
plt.show()
```
